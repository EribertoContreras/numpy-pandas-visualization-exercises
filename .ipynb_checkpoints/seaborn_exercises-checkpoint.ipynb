4) Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
#chipotle = data('chipotle')
from env import get_db_url
url = get_db_url('chipotle')
```

```python
chipotle = pd.read_sql("SELECT * FROM orders", url)
chipotle
```

```python
chipotle['total_price'] = (chipotle.item_price.str.replace('$', '')).astype(float)
```

```python
top_4_items = chipotle.groupby('item_name').quantity.sum().nlargest(4)
top_4_items
```

```python
top_4_revenue = chipotle[chipotle.item_name.isin(top_4_items.index)].groupby('item_name').total_price.sum().reset_index()
top_4_revenue
```

```python
sns.barplot(data = top_4_revenue, x = 'item_name', y = 'total_price')
```
